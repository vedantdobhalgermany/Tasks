### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
import math
x1 = int(input("x1:"))
y1 = int(input("y1:"))
x2 = int(input("x2:"))
y2 = int(input("y2:"))
dist = math.sqrt((x2-x1)**2+(y2-y1)**2)
print(dist)
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
#Input
print("Enter the details :")
head = int(input("Heads :"))
legs = int(input("Legs :"))

# Equations and formula
c = (4*head - legs)//2 #chicken
d = head - c   #dogs

# Output

print("Total Chickens :",c)
print("Total Dogs :",d)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Input
n = int(input("Enter n :"))

#formula
sum = (n*(n+1)*(2*n+1))/6

#output
print("Result :",sum)
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Input for terms
t1 = int(input("Term 1 :"))
t2 = int(input("Term 2 :"))

#Input for nth term
n = int(input("Enter the value of n "))

# Calculations
d = t2 - t1

# Formula
an = t1 + (n-1)*d

# Output
print("nth term = ",an)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# inputs for fraction 1
print("Enter the values of first fraction ")
n1 = int(input("Numerator "))
d1 = int(input("Denominator "))

# inputs for fraction 2
print("Enter the values of second fraction ")
n2 = int(input("Numerator "))
d2 = int(input("Denominator "))

# calculations
nx = (n1*d2)+(n2*d1)
ny = d1*d2

# formula
sum = nx/ny

# output
print("Sum :",sum)
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# Inputs for tank
print("Enter the values for the tank")

L = int(input("Length :"))
B = int(input("Breadth :"))
H = int(input("Height :"))

# Inputs for glass
print("Enter the values for glass")

h = int(input("Height :"))
r =int(input("Radius :"))

# Calculations
vol_tank = L*B*H
vol_glass = 3.14*r*r*h

# Formula
total = vol_tank/vol_glass

# Output
print("Total number of glasses of milk that can be taken out is ",total)
````
