### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Input
sal = int(input("Enter salary in LPA: "))

# Tax Deduction
if sal < 5:
    sal = sal
elif sal >5 and sal < 10:
    sal -= 0.1*sal
elif sal >10 and sal < 20:
    sal -= 0.2*sal
elif sal >20:
    sal -= 0.3*sal

# Deduction on CAT logic
new_sal = sal - 0.1*sal
new_sal -= 0.05*new_sal
new_sal -= 0.03*new_sal


print("In-hand salary: ",int(new_sal))
````

### `Problem 4`: Write a menu-driven program -
1. cm to ft
2. km to miles
3. USD to INR
4. exit

````python
print("Enter the choice :")
print("1. cm to ft")
print("2. km to miles")
print("3. USD to INR")
print("4. Exit")

while True:
    choice = int(input("Enter your choice :"))
    if choice == 1:
        cm = int(input("Enter the value \"cm\":"))
        val = cm/30.48
        print("Value in ft: ", val)
    elif choice == 2:
        km = int(input("Enter the value \"km\":"))
        val = km*0.621371
        print("Value in miles: ", val)
    elif choice == 3:
        USD = int(input("Enter the value \"USD\":"))
        val = USD*83.5
        print("Value in INR: ", val)
    elif choice == 4:
        print("Exit")
        break

````

### `Problem 5` - Exercise 12: Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

````python
a = 0
b = 1
n = int(input("Enter the total number of terms :"))

i = 0
while i<n:
    if i == 0:
        print(a, end=" ")
    elif i == 1:
        print(b, end=" ")
    else:
        next = a + b
        a = b
        b = next
        print(next, end = ' ')
    i += 1
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
n = int(input("Enter the number for factorial : "))

i = 1
fact = 1
if n == 0:
    print("Factorial of 0 is 1")
else:
    while i<=n:
        fact = fact * i
        i += 1
    print(fact)
````

### `Problem 7` - Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
n = int(input("Number:"))

rev = 0
while n>0:
    digit = n%10
    rev = rev * 10 + digit
    n = n//10
print(rev)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```
